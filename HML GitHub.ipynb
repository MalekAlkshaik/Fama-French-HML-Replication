```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import datetime as dt
from pandas.tseries.offsets import MonthEnd
import numpy as np
from dateutil.relativedelta import relativedelta
import MalekFinance as mf
```

```python
data = mf.read('WRDS Data','Compustat US Annual Fundamental + EBIT Dara',index_col=1,parse_dates=True,dayfirst=True)
data['date'] = data.index
data.drop(['prcc_c'],axis=1,inplace=True)
data = data.groupby(['GVKEY', data.index]).last()
data.reset_index(level='GVKEY',inplace=True)
data = data[['GVKEY','csho','ebit','seq','gp','at','prcc_f']]
compudata = mf.read('WRDS Data','CRSP Compustat Monthly Returns & Size Data 63-22',index_col=9,parse_dates=True,dayfirst=True)
compudata['date'] = compudata.index
compudata = compudata.groupby(['GVKEY', compudata.index]).last()
compudata.reset_index(level='GVKEY',inplace=True)
compudata['Adj Close'] = ((compudata['prccm']/compudata['ajexm'])*compudata['trfm'])
compudata['RET'] = compudata.groupby('GVKEY')['Adj Close'].pct_change()
compudata['LPERMNO'] = compudata['LPERMNO'].astype('str')
compudata['LPERMNO'] = compudata['LPERMNO'].astype('object')
column_conversion = dict(zip(compudata['LPERMNO'],compudata['GVKEY']))
```

```python
data = data[data.GVKEY.isin(compudata.GVKEY)]
compudata = compudata[compudata.GVKEY.isin(data.GVKEY)]
NYSE = compudata.query('exchg == 11')
NotNYSE = compudata.query('exchg != 11')
dataNYSE = data[data.GVKEY.isin(NYSE.GVKEY)]
dataNotNYSE = data[data.GVKEY.isin(NotNYSE.GVKEY)]
joinerNYSE = NYSE[['GVKEY','prccm','cshoq']]
joinerNYSE = joinerNYSE.groupby('GVKEY').resample('Q').last()
joinerNYSE = joinerNYSE.reset_index(level='GVKEY',drop=True)
joinerNYSE['Cap'] = joinerNYSE['prccm'] * joinerNYSE['cshoq']
joinerNotNYSE = NotNYSE[['GVKEY','prccm','cshoq']]
joinerNotNYSE = joinerNotNYSE.groupby('GVKEY').resample('Q').last()
joinerNotNYSE = joinerNotNYSE.reset_index(level='GVKEY',drop=True)
joinerNotNYSE['Cap'] = joinerNotNYSE['prccm'] * joinerNotNYSE['cshoq']
dataNYSE = dataNYSE.groupby('GVKEY').resample('Y').last()
dataNYSE = dataNYSE.reset_index(level='GVKEY',drop=True)
dataNotNYSE = dataNotNYSE.groupby('GVKEY').resample('Y').last()
dataNotNYSE = dataNotNYSE.reset_index(level='GVKEY',drop=True)
NyseCapsJune = joinerNYSE[joinerNYSE.index.month == 6]
NyseCapsDecember = joinerNYSE[joinerNYSE.index.month == 12]
NotNyseCapsDecember = joinerNotNYSE[joinerNotNYSE.index.month == 12]
NyseCapsJune.reset_index(inplace=True)
NyseCapsDecember.reset_index(inplace=True)
NotNyseCapsDecember.reset_index(inplace=True)
NyseCapsJune = NyseCapsJune[~NyseCapsJune.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
NyseCapsJune = NyseCapsJune.pivot(index='datadate',columns='GVKEY',values='Cap')
NyseCapsDecember = NyseCapsDecember[~NyseCapsDecember.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
NyseCapsDecember = NyseCapsDecember.pivot(index='datadate',columns='GVKEY',values='Cap')
NotNyseCapsDecember = NotNyseCapsDecember[~NotNyseCapsDecember.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
NotNyseCapsDecember = NotNyseCapsDecember.pivot(index='datadate',columns='GVKEY',values='Cap')
Caps_December_All = pd.concat([NotNyseCapsDecember,NyseCapsDecember],axis=1)
dataNYSE.reset_index(inplace=True)
dataNYSE = dataNYSE[~dataNYSE.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
dataNotNYSE.reset_index(inplace=True)
dataNotNYSE = dataNotNYSE[~dataNotNYSE.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
Book_Value_NYSE = dataNYSE.pivot(index='datadate',columns='GVKEY',values='seq')
Book_Value_NotNYSE = dataNotNYSE.pivot(index='datadate',columns='GVKEY',values='seq')
Book_Value_All = pd.concat([Book_Value_NotNYSE,Book_Value_NYSE],axis=1)
compudata.reset_index(inplace=True)
compudata = compudata[~compudata.duplicated(subset=['GVKEY', 'datadate'], keep='first')]
Returns_df = compudata.pivot(index='datadate', columns='GVKEY', values='RET')
```

```python
frames = [Caps_December_All,NyseCapsDecember,NotNyseCapsDecember,NyseCapsJune,Book_Value_All,Book_Value_NYSE,Book_Value_NotNYSE]
def transformation(x):
    x.drop([np.nan],axis=1,inplace=True)
    x.columns = x.columns.astype('int')    
    x = x.apply(pd.to_numeric,errors='coerce')
for i in range(len(frames)):
    transformation(frames[i])
Caps_December_All,NyseCapsDecember,NotNyseCapsDecember,NyseCapsJune,Book_Value_All,Book_Value_NYSE,Book_Value_NotNYSE = frames
Returns_df.columns = Returns_df.columns.astype('int')
Returns_df = Returns_df.apply(pd.to_numeric,errors='coerce')
formation_dates = pd.date_range('1967-12','2020-12-31',freq='Y') + MonthEnd(6)
return_dates = pd.date_range('1968-7','2022-06-30',freq='M')
```

```python
def HML(formation,offset):
    lagged_formation = formation - MonthEnd(6)
    NYSE_Size_Formation = pd.DataFrame(NyseCapsJune.loc[formation]).dropna()
    NYSE_Size_Lagged_Formation = pd.DataFrame(NyseCapsDecember.loc[lagged_formation])
    All_Size_Lagged_Formation = pd.DataFrame(Caps_December_All.loc[lagged_formation])
    Book_Value_All_Lagged_Formation = pd.DataFrame(Book_Value_All.loc[lagged_formation])
    Book_Value_NYSE_Lagged_Formation = pd.DataFrame(Book_Value_NYSE.loc[lagged_formation])
    Good_Data = Book_Value_All_Lagged_Formation.merge(All_Size_Lagged_Formation,right_index=True,left_index=True,how='inner').dropna()
    NYSE_Breakpoint = NYSE_Size_Formation[NYSE_Size_Formation.index.isin(Good_Data.index)].median()[0]
    merged = Book_Value_NYSE_Lagged_Formation.merge(NYSE_Size_Lagged_Formation,left_index=True,right_index=True,how='inner').dropna()
    Returns_Check_NYSE = Returns_df.loc[formation + MonthEnd(offset),Returns_df.columns.isin(merged.index)].dropna()
    merged = merged.merge(Returns_Check_NYSE,left_index=True,right_index=True,how='inner')
    merged['B/M'] = merged.iloc[:,0] / merged.iloc[:,1]
    merged = merged[merged['B/M'] > 0]
    Value_Threshold = merged['B/M'].quantile(0.7)
    Growth_Threshold = merged['B/M'].quantile(0.3)
    Good_Data['B/M'] = Good_Data.iloc[:,0] / Good_Data.iloc[:,1]
    Good_Data = Good_Data[Good_Data['B/M'] > 0]
    Returns_Check_All = Returns_df.loc[formation + MonthEnd(offset),Returns_df.columns.isin(Good_Data.index)].dropna()
    Good_Data = Good_Data.merge(Returns_Check_All,left_index=True,right_index=True,how='inner').dropna()
    Good_Data.columns = ['Book','Size','B/M','Forward Return']
    Small = Good_Data[Good_Data.iloc[:,1] < NYSE_Breakpoint]
    Large = Good_Data[Good_Data.iloc[:,1] > NYSE_Breakpoint]
    Large_Value = Large[Large['B/M'] > Value_Threshold]
    Large_Growth = Large[Large['B/M'] < Growth_Threshold]
    Small_Value = Small[Small['B/M'] > Value_Threshold]
    Small_Growth = Small[Small['B/M'] < Growth_Threshold]
    Large_Value['Weight'] = Large_Value['Size']/sum(Large_Value['Size'])
    Large_Growth['Weight'] = Large_Growth['Size']/sum(Large_Growth['Size'])
    Small_Value['Weight'] = Small_Value['Size']/sum(Small_Value['Size'])
    Small_Growth['Weight'] = Small_Growth['Size']/sum(Small_Growth['Size'])
    Large_Value_RET = sum(Large_Value['Forward Return'] * Large_Value['Weight'])
    Large_Growth_RET = sum(Large_Growth['Forward Return'] * Large_Growth['Weight'])
    Small_Value_RET = sum(Small_Value['Forward Return'] * Small_Value['Weight'])
    Small_Growth_RET = sum(Small_Growth['Forward Return'] * Small_Growth['Weight'])
    Portfolio_Return = ((Large_Value_RET + Small_Value_RET) - (Large_Growth_RET + Small_Growth_RET))/2
    return Portfolio_Return

portfolioreturns = {i: [HML(date, i) for date in formation_dates] for i in range(1, 13)}
shifted_dates_lists = [
    [date + relativedelta(months=num_months) for date in formation_dates]
    for num_months in range(13)]
dataframes = {
    i: pd.DataFrame(data=portfolioreturns[i], index=shifted_dates_lists[i])
    for i in range(1, 13)}
df = pd.concat(dataframes.values(), axis=1)
HML_Returns = pd.DataFrame({'HML': df.sum(axis=1)}).resample('M').last()
HML_Returns.loc['1993-01-31'] = 0
HML_Returns.sort_index(inplace=True)
```
